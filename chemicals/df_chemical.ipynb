```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
```

```python
df = pd.read_csv('dfnitro.csv', index_col = 0)
```

```python
#visualize
columns = df.columns

for col in columns:
    plt.scatter(df['TOTAL NITROGEN mg/l'], df[col])
    plt.title(col)
    plt.show()
```
